#K-Means Clustering

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
x,y=make_blobs(n_samples=1000,centers=3,n_features=2,random_state=23)
```

```python
y
```

```python
sns.scatterplot(x=x[:,0],y=x[:,1],hue=y, palette='viridis')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=23)
```

```python
from sklearn.cluster import KMeans
```

```python
  wcss=[]

  for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++')
```

```python
y_labels=kmeans.fit_predict(x_train)
y_labels
```

```python
sns.scatterplot(x=x_train[:,0],y=x_train[:,1],hue=y_labels, palette='viridis')
```

```python
ts_labels=kmeans.predict(x_test)
sns.scatterplot(x=x_test[:,0],y=x_test[:,1],hue=ts_labels, palette='viridis')
```

**Performance Matrix**

```python
from sklearn.metrics import silhouette_score
```

```python
sil=[]
for k in range(2,11):
  kmeans=KMeans(n_clusters=k,init='k-means++')
  kmeans.fit(x_train)
  sc=silhouette_score(x_train,kmeans.labels_)
  sil.append(sc)
```

```python
sil
```

```python
plt.plot(range(2,11),sil)
plt.xticks(range(2,11))
plt.xlabel("Number of clusters")
plt.ylabel("Silhoutte coefficient")
plt.show()
```

#Hierarchical clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram ,linkage
```

```python
ds=datasets.load_iris()
```

```python
df=pd.DataFrame(ds.data,columns=ds.feature_names)
```

```python
df
```

```python
sc=StandardScaler()
```

```python
x=sc.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca_sc=pca.fit_transform(x)
```

```python
plt.scatter(pca_sc[:,0],pca_sc[:,1])
```

```python
plt.figure(figsize=(20,7))
plt.title('Dendogram')

dendrogram(linkage(pca_sc,method='ward'))
plt.title('Dendogram')
plt.xlabel("Sample Index")
plt.ylabel("Eucledian Distance")
```

```python
cluster=AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
cluster.fit(pca_sc)
```

```python
plt.scatter(pca_sc[:,0],pca_sc[:,1],c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
sil=[]
for k in range(2,11):
  a=AgglomerativeClustering(n_clusters=k,metric='euclidean',linkage='ward')
  a.fit(x_train)
  sc=silhouette_score(x_train,a.labels_)
  sil.append(sc)
```

```python

plt.plot(range(2,11),sil)
plt.xticks(range(2,11))
plt.xlabel("Number of clusters")
plt.ylabel("Silhoutte coefficient")
plt.show()
```

#DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x,y=make_moons(n_samples=1000,noise=0.05)
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
x_st=sc.fit_transform(x)
```

```python
db=DBSCAN(eps=0.04)
db.fit(x_st)
plt.scatter(x_st[:,0],x_st[:,1],c=db.labels_)
```

#Isolation Anomaly Detection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
df=pd.read_csv('/content/diabetes.csv')
```

```python
df.head()
```

```python
df=df.drop(df.iloc[:,2:7],axis=1)
df=df.drop(columns='Pregnancies',axis=1)
df=df.drop(columns='Outcome',axis=1)
```

```python
df.head()
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
cl=IsolationForest(contamination=0.05)
cl.fit(df)
pre=cl.predict(df)
```

```python
pre
```

```python
ind=np.where(pre<0)
ind
```

```python
x=df.values
```
